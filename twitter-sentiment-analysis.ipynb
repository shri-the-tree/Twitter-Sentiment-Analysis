```python
data=data[['text','label']]
```

```python
data['label'][data['label']==4]=1
```

```python
data_pos = data[data['label'] == 1]
data_neg = data[data['label'] == 0]
```

```python
data_pos = data_pos.iloc[:int(20000)]
data_neg = data_neg.iloc[:int(20000)]
```

```python
data = pd.concat([data_pos, data_neg])
```

```python
data['text']=data['text'].str.lower()
```

```python
data['text'].tail()
```

```python
stopwords_list = stopwords.words('english')
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def cleaning_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
data['text'] = data['text'].apply(lambda text: cleaning_stopwords(text))
data['text'].head()
```

```python
english_punctuations = string.punctuation
punctuations_list = english_punctuations
def cleaning_punctuations(text):
    translator = str.maketrans('', '', punctuations_list)
    return text.translate(translator)
```

```python
data['text']= data['text'].apply(lambda x: cleaning_punctuations(x))
data['text'].tail()
```

```python
def cleaning_repeating_char(text):
    return re.sub(r'(.)\1+', r'\1', text)
```

```python
data['text'] = data['text'].apply(lambda x: cleaning_repeating_char(x))
data['text'].tail()
```

```python
def cleaning_email(data):
    return re.sub('@[^\s]+', ' ', data)
```

```python
data['text']= data['text'].apply(lambda x: cleaning_email(x))
data['text'].tail()
```

```python
def cleaning_URLs(data):
    return re.sub('((www\.[^\s]+)|(https?://[^\s]+))',' ',data)
```

```python
data['text'] = data['text'].apply(lambda x: cleaning_URLs(x))
data['text'].tail()
```

```python
def cleaning_numbers(data):
    return re.sub('[0-9]+', '', data)
```

```python
data['text'] = data['text'].apply(lambda x: cleaning_numbers(x))
data['text'].tail()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
data['text'] = data['text'].apply(tokenizer.tokenize)
```

```python
data['text'].head()
```

```python
st = nltk.PorterStemmer()
def stemming_on_text(data):
    text = [st.stem(word) for word in data]
    return data

data['text']= data['text'].apply(lambda x: stemming_on_text(x))
```

```python
data['text'].head()
```

```python
lm = nltk.WordNetLemmatizer()
def lemmatizer_on_text(data):
    text = [lm.lemmatize(word) for word in data]
    return data

data['text'] = data['text'].apply(lambda x: lemmatizer_on_text(x))
```

```python
data['text'].head()
```

```python
X=data.text
y=data.label
```

```python
max_len = 500
tok = Tokenizer(num_words=2000)
tok.fit_on_texts(X)
sequences = tok.texts_to_sequences(X)
sequences_matrix = sequence.pad_sequences(sequences,maxlen=max_len)
```

```python
sequences_matrix.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(sequences_matrix, y, test_size=0.3, random_state=2)
```

```python
def tensorflow_based_model(): 
    inputs = Input(name='inputs',shape=[max_len])
    layer = Embedding(2000,50,input_length=max_len)(inputs) 
    layer = LSTM(64)(layer) 
    layer = Dense(256,name='FC1')(layer) 
    layer = Activation('relu')(layer) 
    layer = Dropout(0.5)(layer) 
    layer = Dense(1,name='out_layer')(layer) 
    layer = Activation('sigmoid')(layer) 
    model = Model(inputs=inputs,outputs=layer) 
    return model 
```

```python
model = tensorflow_based_model() 
model.compile(loss='binary_crossentropy',optimizer=RMSprop(),metrics=['accuracy'])  
```

```python
history=model.fit(X_train,Y_train,batch_size=80,epochs=6, validation_split=0.1)# here we are starting the training of model by feeding the training data
print('Training finished !!')
```

```python
accr1 = model.evaluate(X_test,Y_test) 
```

```python
print('Test set\n  Accuracy: {:0.2f}'.format(accr1[1]))
```

```python
y_pred = model.predict(X_test) 
y_pred = (y_pred > 0.5) 
```

```python
print('\n')
print("confusion matrix")
print('\n')
CR=confusion_matrix(Y_test, y_pred)
print(CR)
print('\n')

fig, ax = plot_confusion_matrix(conf_mat=CR,figsize=(10, 10),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
plt.show()
```

```python
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.models import Sequential
model_bilstm = Sequential()
model_bilstm.add(Embedding(input_dim=5000, output_dim=128, input_length=20))
model_bilstm.add(Bidirectional(LSTM(64)))
model_bilstm.add(Dense(1, activation='sigmoid'))
```

```python
model_bilstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_bilstm = model_bilstm.fit(X_train, Y_train, epochs=5, validation_split=0.2, batch_size=32)
```

```python
loss_bilstm, accuracy_bilstm = model_bilstm.evaluate(X_test, Y_test)
print(f'BiLSTM Test Accuracy: {accuracy_bilstm}')
```

```python
from tensorflow.keras.layers import GRU
model_gru = Sequential()
model_gru.add(Embedding(input_dim=5000, output_dim=128, input_length=20))
model_gru.add(GRU(64))
model_gru.add(Dense(1, activation='sigmoid'))
```

```python
model_gru.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_gru = model_gru.fit(X_train, Y_train, epochs=5, validation_split=0.2, batch_size=32)
```

```python
loss_gru, accuracy_gru = model_gru.evaluate(X_test, Y_test)
print(f'GRU Test Accuracy: {accuracy_gru}')
```
